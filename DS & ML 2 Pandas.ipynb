```python
import pandas as pd

# Create a sample DataFrame
data = {
    'animal': ['Dog', 'Dog', 'Cat', 'Cat', 'Dog', 'Cat', 'Dog', 'Cat'],
    'age': [3, 5, 2, 4, 7, 3, 2, 1],
    'weight': [20, 25, 8, 10, 30, 9, 18, 7],
    'color': ['Brown', 'Black', 'Orange', 'White', 'Brown', 'Orange', 'White', 'Black']
}
df = pd.DataFrame(data)
print(df)
```

```python
average_values = df.groupby('animal')[['age', 'weight']].mean()
print(average_values)
```

```python
total_values = df.groupby(['animal', 'color'])[['age', 'weight']].sum()
print(total_values)
```

```python
aggregated_values = df.groupby('animal').agg(
    {'age': 'mean', 'weight': ['min', 'max']}
)
print(aggregated_values)
```

```python
grouped_data = df.groupby('animal')
for group_name, group_data in grouped_data:
    print(f"Animal: {group_name}")
    print(group_data)
```

```python
average_values_no_index = df.groupby('animal', as_index=False)[['age', 'weight']].mean()
print(average_values_no_index)
```

```python
import timeit 
  
# code snippet to be executed only once 
setup = '''
import pandas as pd
import numpy as np
messy_names = pd.Series(["andrew  ","bo;bo","  claire  "])
def cleanup(name):
    name = name.replace(";","")
    name = name.strip()
    name = name.capitalize()
    return name
'''
  
# code snippet whose execution time is to be measured 
stmt_pandas_str = ''' 
messy_names.str.replace(";","").str.strip().str.capitalize()
'''

stmt_pandas_apply = '''
messy_names.apply(cleanup)
'''

stmt_pandas_vectorize='''
np.vectorize(cleanup)(messy_names)
'''
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_pandas_str, 
                    number = 10000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_pandas_apply, 
                    number = 10000) 
```

```python
timeit.timeit(setup = setup, 
                    stmt = stmt_pandas_vectorize, 
                    number = 10000) 
```
